```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/amazon_data_webpage.html"
page=requests.get(url)
soup=BeautifulSoup(page.text,"html5lib")
```

```python
soup.title.text
```

```python
soup.find_all("table")
```

```python
table=soup.find("table",class_=
                {"W(100%) M(0)"})
```

```python
#table=soup.find_all("div", class_="Pb(10px) Ovx(a) W(100%)")
columns=["Date", "Open", "High", "Low", "Close", "Adj close", "Volume"]
amazon_data = pd.DataFrame(columns=columns)

Date =""
Open =""
High =""
Low = ""
Close =""
Adj_close =""
Volume =""
i=1
for tr in table.find_all("tr"):
    i=0
    for td in tr.find_all("td"):
        i=i+1
        if (i==1):
            Date=td.text.replace("\n","")
        if (i==2):
            Open=td.text.replace("\n","")
        if(i==3):
            High=td.text.replace("\n","")
        if(i==4):
            Low=td.text.replace("\n","")
        if(i==5):
            Close=td.text.replace("\n","")
        if(i==6):
            Adj_close=td.text.replace("\n","")
        if(i==7):
            Volume=td.text.replace("\n","")

    if (Date!=""):
            amazon_data=amazon_data.append(
                pd.Series([Date,
                           Open,
                           High,
                           Low,
                           Close,
                           Adj_close,
                           Volume]
                            ,index=columns)
                         ,ignore_index=True)

amazon_data
    
```
