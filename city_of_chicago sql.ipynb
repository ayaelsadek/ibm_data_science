```python
DF=pd.read_sql_query("SELECT TOP 5 \
community_area_name, percent_households_below_poverty \
FROM CENSUS_DATA ORDER BY percent_households_below_poverty DESC;",con)
DF
```

```python
DF=pd.read_sql_query("SELECT community_area_number, COUNT(community_area_number) AS 'frequency'\
FROM CHICAGO_CRIME_DATA \
GROUP BY community_area_number \
ORDER BY COUNT(community_area_number) DESC;",con)
DF
```

```python
DF=pd.read_sql_query("SELECT community_area_name FROM CENSUS_DATA \
WHERE hardship_index = (SELECT MAX(hardship_index) FROM CENSUS_DATA);",con)
DF
```

```python
DF=pd.read_sql_query("SELECT TOP 1 COMMUNITY_AREA_NAME FROM CENSUS_DATA \
WHERE community_area_number=\
(SELECT TOP 1 COMMUNITY_AREA_NUMBER\
 FROM CHICAGO_CRIME_DATA c \
 GROUP BY c.COMMUNITY_AREA_NUMBER\
 ORDER BY COUNT(c.COMMUNITY_AREA_NUMBER)DESC);",con)
DF
```
