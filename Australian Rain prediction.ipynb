```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.metrics import jaccard_score
from sklearn.metrics import f1_score
from sklearn.metrics import log_loss
from sklearn.metrics import confusion_matrix, accuracy_score
import sklearn.metrics as metrics
```

```python
path='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillUp/labs/ML-FinalAssignment/Weather_Data.csv'
data=pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillUp/labs/ML-FinalAssignment/Weather_Data.csv'
)
data.head()
```

```python
df=pd.get_dummies(data=data, columns=["WindGustDir","WindDir9am","WindDir3pm","RainToday"])
df.head()
```

```python
df.replace(['No', 'Yes'], [0,1], inplace=True)
```

```python
df.drop('Date',axis=1,inplace=True)
df=df.astype(float)
features=df.drop(columns="RainTomorrow", axis=1)
y=df["RainTomorrow"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, y, test_size=0.2, random_state=10)
```

```python
LinearReg =LinearRegression().fit(x_train, y_train)

LinearReg
```

```python
predictions =LinearReg.predict(x_test)
predictions 
```

```python
from sklearn.metrics import mean_absolute_error as mae 
from sklearn.metrics import mean_squared_error as mse 
from sklearn.metrics import r2_score 

LinearRegression_MAE = mae(y_test,predictions)
print("LinearRegression_MAE: ",str(LinearRegression_MAE))

LinearRegression_MSE = mse(y_test,predictions)
print("LinearRegression_MSE: ",str(LinearRegression_MSE))

LinearRegression_R2 = r2_score(y_test,predictions)
print("LinearRegression_R2: ",str(LinearRegression_R2))
```

```python
Report = {"Metrics":["MAE","MSE","R2"],"Result":[LinearRegression_MAE,LinearRegression_MSE,LinearRegression_R2]}
pd.DataFrame(Report)
```

```python
KNN=KNeighborsClassifier(n_neighbors=4).fit(x_train, y_train)
```

```python
predictions2 = KNN.predict(x_test)
predictions2
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import jaccard_score
from sklearn.metrics import f1_score

KNN_Accuracy_Score =accuracy_score(y_test,predictions2)
print("KNN_Accuracy_Score: ",str(KNN_Accuracy_Score))

KNN_JaccardIndex = jaccard_score(y_test,predictions2)
print("KNN_JaccardIndex: ",str(KNN_JaccardIndex))

KNN_F1_Score = f1_score(y_test,predictions2)
print("KNN_F1_Score: ",str(KNN_F1_Score))
```

```python
report={'matrics':["accuracy_score","jaccard_score","f1_score"],'result':[KNN_Accuracy_Score, KNN_JaccardIndex, KNN_F1_Score]}
pd.DataFrame(report)
```

```python
Tree =DecisionTreeClassifier().fit(x_train, y_train)
```

```python
predictions3=Tree.predict(x_test)
```

```python
Tree_Accuracy_Score = accuracy_score(y_test,predictions3)
print("Tree_Accuracy_Score: ",str(Tree_Accuracy_Score))

Tree_JaccardIndex = jaccard_score(y_test,predictions3)
print("Tree_JaccardIndex: ",str(Tree_JaccardIndex))

Tree_F1_Score = f1_score(y_test,predictions3)
print("Tree_F1_Score: ",str(Tree_F1_Score))
```

```python
report={'matrics':['accuracy_score', 'jaccard_score', 'f1_score'],'result':[Tree_Accuracy_Score,Tree_JaccardIndex,Tree_F1_Score]}
pd.DataFrame(report)
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, y, test_size=0.2, random_state=1)
```

```python
LR=LogisticRegression(solver='liblinear').fit(x_train, y_train)
```

```python
predictions4=LR.predict(x_test)
predict_proba = LR.predict_proba(x_test)
```

```python
LR_Accuracy_Score = accuracy_score(y_test,predictions4)
print("LR_Accuracy_Score: ",str(LR_Accuracy_Score))

LR_JaccardIndex = jaccard_score(y_test,predictions4)
print("LR_JaccardIndex: ",str(LR_JaccardIndex))


LR_F1_Score =f1_score(y_test,predictions4)
print("LR_F1_Score: ",str(LR_F1_Score)) 

from sklearn.metrics import log_loss
LR_Log_Loss = log_loss(y_test, predict_proba)
print("LR_Log_Loss: ",str(LR_Log_Loss)) 
```

```python
SVM = svm.SVC().fit(x_train, y_train)
```

```python
predictions5 = SVM.predict(x_test)
```

```python
SVM_Accuracy_Score = accuracy_score(y_test,predictions5)
print("SVM_Accuracy_Score: ",str(SVM_Accuracy_Score))

SVM_JaccardIndex = jaccard_score(y_test,predictions5)
print("SVM_JaccardIndex: ",str(SVM_JaccardIndex))

SVM_F1_Score = f1_score(y_test,predictions5)
print("SVM_F1_Score: ",str(SVM_F1_Score)) 
```

```python
report={'matrics':['accuracy_score', 'jaccard_score', 'f1_score'],'result':[SVM_Accuracy_Score,SVM_JaccardIndex,SVM_F1_Score]}
pd.DataFrame(report)
```
